```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dta['tfopwg_disp'] = le.fit_transform(dta['tfopwg_disp'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dta.drop(columns = ['tfopwg_disp'])
Y = dta['tfopwg_disp']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

Models

```python
from sklearn import metrics
```

```python
from sklearn.linear_model import LogisticRegression
#model= LogisticRegression(solver='lbfgs')
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
prediction=model.predict(X_test)
prediction
```

```python
Logi_acc=metrics.accuracy_score(prediction,Y_test)*100
print('The accuracy of the LogisticRegression is',round(Logi_acc,2))
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(Y_test,prediction)
print(cm)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ax=plt.axes()
data_cm=cm
sns.heatmap(data_cm,annot=True,annot_kws={"size": 20}, fmt='d',cmap="Blues", ax = ax )
ax.set_title('Confusion Matrix')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
mymodel=DecisionTreeClassifier()
mymodel.fit(X_train,Y_train)
```

```python
prediction=mymodel.predict(X_test)
prediction
```

```python
Dec_acc=metrics.accuracy_score(prediction,Y_test)*100
print('The accuracy of the Decision Tree is',round(Dec_acc,2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import numpy as np
error_rate = []
for i in range(1,80):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != Y_test))

plt.figure(figsize=(5,15))
plt.plot(range(1,80),error_rate,color='blue', linestyle='dashed',
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
dta1 = dta.drop('tfopwg_disp', axis=1)
```

```python
# Create a scaler object
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(dta1)
```

```python
# Create a PCA instance with 2 components
pca = PCA(n_components=12)

# Fit and transform the scaled data to get the principal components
principal_components = pca.fit_transform(scaled_data)

# Create a new DataFrame with the principal components
pca_data = pd.DataFrame(data = principal_components)
```

```python
#print(f"Variance explained by PC18: {pca.explained_variance_ratio_[17]:.2%}")
#print(f"Variance explained by PC2: {pca.explained_variance_ratio_[18]:.2%}")
print(f"Total variance explained by both: {pca.explained_variance_ratio_.sum():.2%}")
```

```python
# 4. Create and display the scree plot
explained_variance = pca.explained_variance_ratio_
n_components = np.arange(1, len(explained_variance) + 1)

plt.figure(figsize=(10, 7))
plt.plot(n_components, explained_variance, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot for PCA')
plt.xlabel('Number of Principal Components')
plt.ylabel('Proportion of Variance Explained')
plt.grid()
plt.show()
```
