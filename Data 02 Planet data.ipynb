```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import numpy as np
error_rate = []
for i in range(1,80):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != Y_test))

plt.figure(figsize=(5,15))
plt.plot(range(1,80),error_rate,color='blue', linestyle='dashed',
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
da1 = da.drop('disposition', axis=1)
```

```python
# Create a scaler object
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(da1)
```

```python
# Create a PCA instance with 2 components
pca = PCA(n_components=12)

# Fit and transform the scaled data to get the principal components
principal_components = pca.fit_transform(scaled_data)

# Create a new DataFrame with the principal components
pca_data = pd.DataFrame(data = principal_components)
```

```python
#print(f"Variance explained by PC18: {pca.explained_variance_ratio_[17]:.2%}")
#print(f"Variance explained by PC2: {pca.explained_variance_ratio_[18]:.2%}")
print(f"Total variance explained by both: {pca.explained_variance_ratio_.sum():.2%}")
```

```python
# 4. Create and display the scree plot
explained_variance = pca.explained_variance_ratio_
n_components = np.arange(1, len(explained_variance) + 1)

plt.figure(figsize=(10, 7))
plt.plot(n_components, explained_variance, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot for PCA')
plt.xlabel('Number of Principal Components')
plt.ylabel('Proportion of Variance Explained')
plt.grid()
plt.show()
```
